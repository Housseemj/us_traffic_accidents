```python
%python
df_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("s3a://s3-us-accidents/US_Accidents_March23.csv")
)

display(df_raw)

def clean_column_name(col_name):
    return (
        col_name.replace(" ", "_")
        .replace(",", "")
        .replace(";", "")
        .replace("{", "")
        .replace("}", "")
        .replace("(", "")
        .replace(")", "")
        .replace("\n", "")
        .replace("\t", "")
        .replace("=", "")
    )

df_clean = df_raw.toDF(*[clean_column_name(c) for c in df_raw.columns])

df_clean.write.format("delta").mode("overwrite").save("dbfs:/delta/bronze")
```

```python
df_bronze = spark.read.format("delta").load("dbfs:/delta/bronze")
display(df_bronze)
```
